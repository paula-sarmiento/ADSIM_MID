# Verification of 1D diffusion


$$\begin{align}
C(x,t) = C_i 
+ (C_0 - C_i) A(x,t)
\end{align}$$

$$
\begin{aligned}
A(x,t) ={}& 
\frac{1}{2}\,
\operatorname{erfc}\!\left(
    \frac{R x - v t}{2\sqrt{D R t}}
\right)
+ \frac{1}{2}\,
\exp\!\left(\frac{v x}{D}\right)
\operatorname{erfc}\!\left(
    \frac{R x + v t}{2\sqrt{D R t}}
\right)
\\[6pt]
&+ 
\frac{1}{2}
\left[
    2 
    + \frac{v(2L - x)}{D}
    + \frac{v^{2} t}{D R}
\right]
\exp\!\left(\frac{v L}{D}\right)
\operatorname{erfc}\!\left(
    \frac{R(2L - x) + v t}{2\sqrt{D R t}}
\right)
\\[6pt]
&\quad
- \left(\frac{v^{2} t}{\pi D R}\right)^{1/2}
\exp\!\left[
    \frac{v L}{D}
    - \frac{R}{4 D t}\,(2L - x + \tfrac{v t}{R})^{2}
\right]
\end{aligned}
$$

Where: <br>
$C_i=$ Homogeneous initial concentration <br>
$C_0=$ Boundary concentration <br>
$x=$ Position <br>
$v=$ gas seepage <br>
$D=$ Diffusion coefficient <br>
$t=$ Time

```python
def A_1D(x, t, v, D, R, L):
    """
    Analytical solution for 1D diffusion-advection with boundary conditions.
    
    Parameters:
    x: Position
    t: Time
    v: Gas seepage velocity
    D: Diffusion coefficient
    R: Retardation factor
    L: Domain length
    """
    term1 = 0.5 * erfc((R * x - v * t) / (2 * np.sqrt(D * R * t)))
    
    term2 = 0.5 * np.exp(v * x / D) * erfc((R * x + v * t) / (2 * np.sqrt(D * R * t)))
    
    term3 = 0.5 * (2 + v * (2*L - x) / D + v**2 * t / (D * R)) * \
            np.exp(v * L / D) * erfc((R * (2*L - x) + v * t) / (2 * np.sqrt(D * R * t)))
    
    term4 = -np.sqrt(v**2 * t / (np.pi * D * R)) * \
            np.exp(v * L / D - R / (4 * D * t) * (2*L - x + v * t / R)**2)
    
    return term1 + term2 + term3 + term4
```

```python
# Define x points
x_points = np.linspace(0, 1, 100)  # from 0 to 1 m

#final time simulated
t_final = 14400  # seconds
# Diffusion coefficient
D = 1.6e-5   # m^2/s
# Gas seepage velocity
v = 8.5e-6  # m/s (set to 0 for pure diffusion)
# Retardation factor
R = 0.5  # dimensionless
# Domain length
L = 1.0  # m
# Boundary concentration
C0 = 45  # mol/m^3
# Initial concentration
C_i = 41  # mol/m^3

# snapshot times to evaluate
snapshot_times = np.linspace(0, t_final, 25)  # from 0s to 7200s

# Calculate analytical solution at snapshot times
analytical_solutions = {}
for t in snapshot_times:
    #for t=0 just return initial condition
    if t==0:
        analytical_solutions[t] = np.full_like(x_points, C_i)
        continue

    A= A_1D(x_points, t, v, D, R, L)
    if A[0]>1 : #linearly correct for deviations in A from max to 0.5 max at bottom
        error_top= A[0]-1
        error_layer= -1*error_top* (x_points) + error_top 
        A_corrected= A - error_layer
    else: 
        A_corrected=A
    analytical_solutions[t] = C_i + (C0 - C_i) * A_corrected

# Save data in DataFrame for easier col 1: x_points col 2- : concentrations at different times
data = {'Position (m)': x_points}
OB_DATA=[]
for t in snapshot_times:
    data['C'] = analytical_solutions[t]
    if t == 0: #fill with initial condition
        data['C'] = np.full_like(x_points, C_i)
    df = pd.DataFrame(data)
    OB_DATA.append(df)
```

```python
#define a color palette the size of the maximum dataset
palette = sns.color_palette("viridis", max(len(snapshot_times), len(ADSIM_DATA)))

#define plot size
width= 83 * 0.0393701 #in mm to inches
AR= 0.6  # aspect ratio height/width
height= width * AR  # in in

gf.PlotAll(OB_DATA, snapshot_times, 'C', 'Position (m)', palete=palette, legendMode=False, hold=1, xlabel=r'Gas concentration [mol/m$^3$]', ylabel='Depth [m]')

gf.PlotAll(ADSIM_DATA, IDs, 'CO2_Concentration', 'arc_length', palete=palette, legendMode=False, hold=1, xlabel=r'Gas concentration [mol/m$^3$]', ylabel='Depth [m]', mode='*', makersize=3, sample_span=100, linewith=1, xsize=width, ysize=height)

#invert y axis
plt.gca().invert_yaxis()
```
